```python
# Step1.
# 必要なライブラリをインポートする。
import matplotlib.pyplot as plt
import pandas as pd
import japanize_matplotlib

# サンプルデータ（リスト）
data = {
    'カテゴリ': ['ビール', '日本酒', 'ワイン'],
    '売り上げ金額（億円）': [1024, 667, 431]
}
df = pd.DataFrame(data)

# Step2.
# カテゴリー列の要素(弁当・デザートなど)のユニークな出現頻度をカウントするメソッドと、
# そのメソッドの実行結果を棒グラフ化するメソッドを組み合わせたコードを記述する。
freq_counts = df['カテゴリ'].value_counts()

df.plot.bar(x='カテゴリ', y='売り上げ金額（億円）')
plt.xlabel('カテゴリ')
plt.ylabel('売り上げ金額（億円）')
plt.title('カテゴリ別売り上げ')

# Step3.
# グラフを表示させる行「plt.show()」を書く
plt.show()

# Step4.
# ある基準列(商品番号列)ごとに要素をグループにまとめサブ列を集計するメソッドと、
# 基本統計量を算出するメソッドを組み合わせたコードを記載する

# 'カテゴリ'列ごとにグループ化して要素の発生頻度をカウント
freq_counts_grouped = df.groupby('カテゴリ').count()
print(freq_counts_grouped)

# 基本統計量を算出するメソッド
summary = df.describe()
print(summary)
```
