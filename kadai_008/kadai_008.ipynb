```python
# Step1.
# 「1. サンプルデータの取得」
# ワインの成分とワイン種別のデータである「Wine recognition dataset」を取得
from sklearn.datasets import load_wine

# インポートしたload_wine関数を実行し、datasetという変数に代入
dataset = load_wine()
dataset

# 「2. サンプルデータの分割」
# 説明変数（ワインの成分）を変数Xに、目的変数（ワインの種別）を変数yに代入します。
X = dataset.data
y = dataset.target

# train_test_split関数をインポート。
from sklearn.model_selection import train_test_split

# test_sizeに0.3し、学習データとテストデータを7:3に分割する。
train_test_split(X, y, test_size=0.3, random_state=5)

# 4つの要素をそれぞれ、変数X_train, X_test, y_train, y_testに代入
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=5)

# Step2.
# 予測モデルとしてランダムフォレストを選択。RandomForestClassifierクラスをインポート
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(random_state=3)

# Step3.
#「4. 予測モデルの学習」
rf_model.fit(X_train, y_train)
y_pred = rf_model.predict(X_test)

#「5. 予測モデルの評価」
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
# 「決定木」での評価：0.9259259259259259
```
